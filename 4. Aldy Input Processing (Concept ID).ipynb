```python
import pandas
import os

# This query represents dataset "All_WGS" for domain "person" and was generated for All of Us Controlled Tier Dataset v6
dataset_02785573_person_sql = """
    SELECT
        person.person_id,
        p_gender_concept.concept_name as gender,
        person.birth_datetime as date_of_birth,
        p_race_concept.concept_name as race,
        p_ethnicity_concept.concept_name as ethnicity,
        p_sex_at_birth_concept.concept_name as sex_at_birth 
    FROM
        `""" + os.environ["WORKSPACE_CDR"] + """.person` person 
    LEFT JOIN
        `""" + os.environ["WORKSPACE_CDR"] + """.concept` p_gender_concept 
            ON person.gender_concept_id = p_gender_concept.concept_id 
    LEFT JOIN
        `""" + os.environ["WORKSPACE_CDR"] + """.concept` p_race_concept 
            ON person.race_concept_id = p_race_concept.concept_id 
    LEFT JOIN
        `""" + os.environ["WORKSPACE_CDR"] + """.concept` p_ethnicity_concept 
            ON person.ethnicity_concept_id = p_ethnicity_concept.concept_id 
    LEFT JOIN
        `""" + os.environ["WORKSPACE_CDR"] + """.concept` p_sex_at_birth_concept 
            ON person.sex_at_birth_concept_id = p_sex_at_birth_concept.concept_id  
    WHERE
        person.PERSON_ID IN (
            SELECT
                distinct person_id  
            FROM
                `""" + os.environ["WORKSPACE_CDR"] + """.cb_search_person` cb_search_person  
            WHERE
                cb_search_person.person_id IN (
                    SELECT
                        person_id 
                    FROM
                        `""" + os.environ["WORKSPACE_CDR"] + """.cb_search_person` p 
                    WHERE
                        has_whole_genome_variant = 1 
                ) 
            )"""

dataset_02785573_person_df = pandas.read_gbq(
    dataset_02785573_person_sql,
    dialect="standard",
    use_bqstorage_api=("BIGQUERY_STORAGE_API_ENABLED" in os.environ),
    progress_bar_type="tqdm_notebook")

dataset_02785573_person_df.head(5)
```

# Load & Process Selected Samples

```python
# load random sample data set
import pandas as pd

## allele output from all sample sets
df = pd.read_csv("/home/jupyter/workspaces/multiallelicgenesinvestigation/sample_id_for_set1234.csv", header = 0, index_col = 0)
selected_sample = list(df['person_id'])
```

```python
## date of birth to age
from datetime import datetime, date
import numpy as np

def age(born):
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
```

```python
## Include only selected samples 
isin_filter = all_person_df['person_id'].isin(selected_sample)
person_df_isin = all_person_df[isin_filter]
person_df_isin = person_df_isin.sort_values(['person_id'])

age_list = []
for i in person_df_isin['date_of_birth']:
    age_list+= [age(i)]

person_df_isin.drop('date_of_birth', axis = 1, inplace = True)
person_df_isin['age'] = age_list
print(person_df_isin.shape)


isin_filter = all_condition_df['person_id'].isin(selected_sample)
condition_df_isin = all_condition_df[isin_filter]
condition_df_isin = condition_df_isin.sort_values(['person_id'])
print(condition_df_isin.shape)
```

# Processing
Skip/Prefer not to answer/.. = "None Indicated"

```python
import warnings
warnings.filterwarnings('ignore')

person_df = person_df_isin[['person_id','sex_at_birth','race','ethnicity','age']]
person_df['race'] = person_df['race'].replace('PMI: Skip', 'None Indicated')
person_df['race'] = person_df['race'].replace('I prefer not to answer', 'None Indicated')


person_df['ethnicity'] = person_df['ethnicity'].replace('PMI: Skip', 'None Indicated')
person_df['ethnicity'] = person_df['ethnicity'].replace('PMI: Prefer Not To Answer', 'None Indicated')
person_df['ethnicity'] = person_df['ethnicity'].replace('What Race Ethnicity: Race Ethnicity None Of These', 'None of these')

person_df['sex_at_birth'] = person_df['sex_at_birth'].replace('PMI: Skip', 'None Indicated')
person_df['sex_at_birth'] = person_df['sex_at_birth'].replace('No matching concept', 'None Indicated')


concept_df = condition_df_isin[['person_id','condition_concept_id','standard_concept_name']]

person_id = list(set(concept_df['person_id']))
concept_id = list(set(concept_df['condition_concept_id']))
```
